```python
re.findall("日本",s)
```

```python
# 特定した場所を表示
[m.span() for m in re.finditer('日本', s)]
```

```python
# 変数に入れる
results = ([m.span() for m in re.finditer('決勝', s)])
print(results)
```

```python
# for loopで結果を表示！
for i in results:
    print(s[i[0]-20:i[1]+20]+"...")
```

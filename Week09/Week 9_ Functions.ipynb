## 5.1 オリジナルの関数
　いきなり「**オリジナルの関数を作ろう**」といわれても、どうしたらいいかもわかりませんし、なぜそんなことをする必要があるのかもわかりませんよね。それではまず、「**どうやって関数を作るのか**」から勉強していきましょう。なぜオリジナルの関数を作る必要があるのかは後ほど説明します。

　シンプルな関数の作り方は以下のように
```
def 関数名():
  処理
```
のような形で、オリジナルの関数を定義することです。実際に作ってみましょう。

````python
def hello():
    print('Hello Reitaku')
````

注意しないといけないのは、`if文`や`for文`の時と同様に、<font color="red">`def`の次の行は**関数ブロックとして字下げします**</font>。そして、**関数は定義をしただけでは実行されません。**<font color="red">関数を実行するには、print()などと同様に、定義した関数名を打ち込んで実行する必要があります</font>。

### 例1

````python
hello()
````

### 解答例

````python
def mysum(a):
  total = 0
  for ii in range(1,a+1):
    total = total + ii
  print(total)

mysum(100)
````

## 関数の中の変数

````python
#関数の実行
2*mysum(100)
````

`2*mysum(100)`を実行すると10100になると思ったのに、エラーになってしまいました。なぜでしょうか？mysum関数では、最後に`print(sum)`で足し算の結果を出力しています。なので、`2*mysum(100)`は`2*print(sum)`を計算しているような感じになるのでエラーになってしまいます。`print(sum)`は数字ではないですから。では、`mysum`関数の中では、`total`に足し算の結果が代入されますから
```
mysum(100)
2*total
```
と実行してみたらどうでしょうか。

````python
mysum(100)
2*total
````

こちらもエラーになりますね。エラーメッセージを読み解くと、**totalという変数はないよ**と書かれています。`mysum`の中では`total`を使っているのになぜでしょうか？

実は、関数の引数や関数の中で使う変数は、<font color="red">**その関数の中だけで有効**なのです（**ローカル変数の独立性**）</font>。では、`mysum`関数で計算した値を取り出したい場合はどうすればよいでしょうか。

# なぜ関数を作成するのか？

数値だけ変えて似たような処理を複数回行いたい場合を考えます。たとえば

1.   1～1000までの合計を求める
2.   1001～2000までの合計を求める
3.   2001～3000までの合計を求める

です。このとき、関数を使わないとfor文やwhile文のプログラムを3つ作成しないといけません。ですが、関数を使えば、リストの和を計算するプログラムを1つ作成しておけば、あとはその関数を3回実行すれば良いだけになります。このような感じで、関数を使うと、プログラムをすっきり書くことができます。





````python
def hello():
    print('Hello Reitaku')
````

````python
hello()
````
